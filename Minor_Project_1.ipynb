```python
df=df.loc[:,['Age','Gender','Driving_License','Previously_Insured','Vehicle_Age','Vehicle_Damage','Response']]
df=df.loc[df['Response']!=2]
df=df[df['Driving_License']==1  ]
df.drop('Driving_License',axis=1,inplace=True)
```

```python
df.shape
```

```python
plt.figure(figsize=(10,6))
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
df['Gender'].value_counts().plot(kind='line')
```

```python
plt.figure(figsize=(10,6))
plt.title('Distribution of Vehicle Age')
plt.xlabel('Vehicle Age')
plt.ylabel('Count')
df['Vehicle_Age'].value_counts().plot(kind='line')
```

```python
plt.figure(figsize=(10,6))
plt.title('Distribution of Vehicle Damage')
plt.xlabel('Vehicle Damage')
plt.ylabel('Count')
df['Vehicle_Damage'].value_counts().plot(kind='line')
```

```python
plt.figure(figsize=(10,6))
plt.title('Distribution of Previously Insured')
plt.xlabel('Previously Insured')
plt.ylabel('Count')
df['Previously_Insured'].value_counts().plot(kind='line')
```

```python
plt.figure(figsize=(10,6))
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
df['Age'].value_counts().plot(kind='bar')
```
